## Sentiment Analysis

```python
# Model
mdlSnt = pipeline('sentiment-analysis')

# Parameters
#prmStatement = 'We are happy to go on vacation this spring break.'
prmStatement = ['This is the best day of my life!',
"I'm really happy with how everything turned out.",
"I'm satisfied with the results, they met my expectations.", #Removed indent
"It was okay, not great but not bad either.",
"I'm a bit disappointed with the outcome",
"This didn't go as well as I had hoped.",
"I'm really unhappy with the way things turned out.",
"This is just terrible; nothing went right.",
"I can't believe how awful this experience was.",
"Everything is a complete disaster, I'm devastated.",
        'I am struggling with stress']
# Predict
mdlSnt(prmStatement)
```

Can perform sentiment analysis in other languages as well using `model = 'nlptown/bert-base-multilingual-uncased-sentiment'`

```python
# Model
mdlSnt = pipeline('sentiment-analysis', model = 'nlptown/bert-base-multilingual-uncased-sentiment')

# Parameters
prmStatement = ' Me cuesta, pero cuanto más me esfuerzo, más progreso' # .Zorlanıyorum ama zorlandıkça gelişiyorum

# Predict
mdlSnt(prmStatement)
```

The result is in stars as opposed to score... this is as per the model definition.  See model details in HF for more details.

## Question Answering

### Example

```python
from transformers import pipeline

# Initialize the question-answering pipeline
mdlQa = pipeline("question-answering")

# Powerlifting text
f = """Powerlifting is a strength sport that consists of three attempts at maximal weight on three lifts: squat, bench press, and deadlift. Unlike Olympic weightlifting, which includes the snatch and clean and jerk, powerlifting focuses on raw maximal strength. Powerlifters often train in specialized programs that emphasize progressive overload, focusing on improving strength over time. Competitions are usually divided into weight classes and age groups, allowing lifters to compete against others of similar size and experience. The sport requires not only physical strength but also mental focus and discipline to achieve personal bests."""

# Questions related to powerlifting
question1 = "What are the three main lifts in powerlifting?"
question2 = "How does powerlifting differ from Olympic weightlifting?"
question3 = "What factors are emphasized in powerlifting training programs?"

# Predict
print("Answer 1:", mdlQa(question=question1, context=f))
print("Answer 2:", mdlQa(question=question2, context=f))
print("Answer 3:", mdlQa(question=question3, context=f))
```

```python
from transformers import pipeline

# Initialize the question-answering pipeline
mdlQa = pipeline("question-answering")

# Techno music text
f = """Techno music is a genre of electronic dance music that emerged in Detroit, Michigan, during the mid-to-late 1980s. It is characterized by a repetitive 4/4 beat, often produced using drum machines, synthesizers, and digital audio workstations. The genre is known for its futuristic sound, which blends elements of electronic music with rhythmic and often minimalist beats. Techno has evolved into various subgenres, including acid techno, minimal techno, and melodic techno. It is commonly played in nightclubs and festivals, with DJs often mixing tracks to create continuous, high-energy sets that keep dancers moving throughout the night."""

# Questions related to techno music
question1 = "Where and when did techno music emerge?"
question2 = "What instruments or tools are commonly used to produce techno music?"
question3 = "In what settings is techno music commonly played?"

# Predict
print("Answer 1:", mdlQa(question=question1, context=f))
print("Answer 2:", mdlQa(question=question2, context=f))
print("Answer 3:", mdlQa(question=question3, context=f))
```
